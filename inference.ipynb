** YOLOv8 PPE Detection Script**

```python
import os
import cv2
import numpy as np
from pathlib import Path
from ultralytics import YOLO
import shutil
import yaml
from sklearn.model_selection import train_test_split

def crop_and_adjust_annotations(image_path, annotation_path, person_detector, output_dir, classes):
    image = cv2.imread(str(image_path))
    height, width = image.shape[:2]

    results = person_detector(image)
    person_boxes = results[0].boxes.xyxy.cpu().numpy()

    with open(annotation_path, 'r') as f:
        annotations = [line.strip().split() for line in f]

    cropped_images = []
    cropped_annotations = []

    for i, person_box in enumerate(person_boxes):
        x1, y1, x2, y2 = map(int, person_box)
        cropped_image = image[y1:y2, x1:x2]
        cropped_height, cropped_width = cropped_image.shape[:2]

        adjusted_annotations = []
        for ann in annotations:
            cls, x_center, y_center, w, h = map(float, ann)
            x_center *= width
            y_center *= height
            w *= width
            h *= height

            if (x1 < x_center < x2) and (y1 < y_center < y2):
                new_x_center = (x_center - x1) / cropped_width
                new_y_center = (y_center - y1) / cropped_height
                new_w = w / cropped_width
                new_h = h / cropped_height

                adjusted_annotations.append(f"{int(cls)} {new_x_center:.6f} {new_y_center:.6f} {new_w:.6f} {new_h:.6f}")

        if adjusted_annotations:
            cropped_images.append(cropped_image)
            cropped_annotations.append(adjusted_annotations)

    for i, (img, anns) in enumerate(zip(cropped_images, cropped_annotations)):
        img_filename = f"{image_path.stem}_person{i}.jpg"
        ann_filename = f"{image_path.stem}_person{i}.txt"

        cv2.imwrite(str(output_dir / 'images' / img_filename), img)
        with open(output_dir / 'labels' / ann_filename, 'w') as f:
            f.write('\n'.join(anns))

def prepare_ppe_dataset(input_dir, output_dir, person_model_path, classes_path):
    person_detector = YOLO(person_model_path)

    input_dir = Path(input_dir)
    output_dir = Path(output_dir)
    (output_dir / 'images').mkdir(parents=True, exist_ok=True)
    (output_dir / 'labels').mkdir(parents=True, exist_ok=True)

    with open(classes_path, 'r') as f:
        classes = [line.strip() for line in f]

    for img_path in input_dir.glob('images/*.jpg'):
        ann_path = input_dir / 'labels' / f"{img_path.stem}.txt"
        if ann_path.exists():
            crop_and_adjust_annotations(img_path, ann_path, person_detector, output_dir, classes)

def create_data_yaml(output_dir, classes_path):
    with open(classes_path, 'r') as f:
        classes = [line.strip() for line in f]

    data = {
        'train': str(output_dir / 'train' / 'images'),
        'val': str(output_dir / 'val' / 'images'),
        'nc': len(classes),
        'names': classes
    }

    with open(output_dir / 'data.yaml', 'w') as f:
        yaml.dump(data, f, default_flow_style=False)

def split_dataset(output_dir, val_split=0.2):
    images_dir = output_dir / 'images'
    labels_dir = output_dir / 'labels'

    all_images = list(images_dir.glob('*.jpg'))
    print(f"Found {len(all_images)} images in {images_dir}")

    if len(all_images) == 0:
        print("No images found. Skipping dataset split.")
        return
    train_images, val_images = train_test_split(all_images, test_size=val_split, random_state=42)

    for split, image_list in [('train', train_images), ('val', val_images)]:
        (output_dir / split / 'images').mkdir(parents=True, exist_ok=True)
        (output_dir / split / 'labels').mkdir(parents=True, exist_ok=True)

        for img_path in image_list:
            shutil.move(str(img_path), str(output_dir / split / 'images' / img_path.name))
            label_path = labels_dir / f"{img_path.stem}.txt"
            if label_path.exists():
                shutil.move(str(label_path), str(output_dir / split / 'labels' / label_path.name))

def train_ppe_detection(data_yaml, epochs=5, imgsz=640, batch=16):
    model = YOLO('yolov8n.yaml')
    results = model.train(
        data=data_yaml,
        epochs=epochs,
        imgsz=imgsz,
        batch=batch,
        name='ppe_detection'
    )
    return results

def main(input_dir, output_dir, person_model_path, classes_path, epochs=100, imgsz=640, batch=16):
    prepare_ppe_dataset(input_dir, output_dir, person_model_path, classes_path)

    split_dataset(Path(output_dir))

    create_data_yaml(Path(output_dir), classes_path)

    results = train_ppe_detection(str(Path(output_dir) / '/content/drive/MyDrive/datasets/dataset.yaml'), epochs, imgsz, batch)

    print("Training completed. Results:")
    print(results)

if __name__ == "__main__":
    import argparse

    parser = argparse.ArgumentParser(description='Train YOLOv8 model for PPE detection')
    parser.add_argument('input_dir', help='Path to the input directory containing images and labels')
    parser.add_argument('output_dir', help='Path to the output directory for processed data')
    parser.add_argument('person_model_path', help='Path to the pre-trained person detection model')
    parser.add_argument('classes_path', help='Path to the text file containing class names')
    parser.add_argument('--epochs', type=int, default=5, help='Number of training epochs')
    parser.add_argument('--imgsz', type=int, default=640, help='Input image size for training')
    parser.add_argument('--batch', type=int, default=16, help='Batch size for training')

    args_list = [
        '/content/drive/MyDrive/datasets/images/train',  
        '/content/drive/MyDrive/datasets/output', 
        '/content/runs/detect/train/weights/best.pt',  
        '/content/drive/MyDrive/datasets/classes.txt'  
    ]
    args = parser.parse_args(args_list)

    main(args.input_dir, args.output_dir, args.person_model_path, args.classes_path, args.epochs, args.imgsz, args.batch)
```

**YOLOv8 Inference Script**

```python
import argparse
import cv2
import os
from ultralytics import YOLO

def draw_boxes(image, boxes, labels, scores):
    for box, label, score in zip(boxes, labels, scores):
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        text = f"{label}: {score:.2f}"
        cv2.putText(image, text, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    return image

def perform_inference(input_dir, output_dir, person_model, ppe_model):
    person_detector = YOLO(person_model)
    ppe_detector = YOLO(ppe_model)

    for image_file in os.listdir(input_dir):
        if image_file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(input_dir, image_file)
            image = cv2.imread(image_path)

            person_results = person_detector(image)

            for person_box in person_results[0].boxes.xyxy:
                x1, y1, x2, y2 = map(int, person_box)
                person_crop = image[y1:y2, x1:x2]
                ppe_results = ppe_detector(person_crop)

                for ppe_box in ppe_results[0].boxes.xyxy:
                    ppe_x1, ppe_y1, ppe_x2, ppe_y2 = map(int, ppe_box)
                    ppe_box[0] += x1
                    ppe_box[1] += y1
                    ppe_box[2] += x1
                    ppe_box[3] += y1

                image = draw_boxes(image, ppe_results[0].boxes.xyxy, ppe_results[0].boxes.cls, ppe_results[0].boxes.conf)

            image = draw_boxes(image, person_results[0].boxes.xyxy, person_results[0].boxes.cls, person_results[0].boxes.conf)

            output_path = os.path.join(output_dir, f"inference_{image_file}")
            cv2.imwrite(output_path, image)

if __name__ == "__main__":
    args = argparse.Namespace(
        input_dir='/content/drive/MyDrive/datasets/images/train',
        output_dir='/content/drive/MyDrive/datasets/output',
        person_det_model='/content/runs/detect/train/weights/best.pt',
        ppe_detection_model='/content/runs/detect/ppe_detection3/weights/best.pt'
    )

    perform_inference(args.input_dir, args.output_dir, args.person_det_model, args.ppe_detection_model)
```
